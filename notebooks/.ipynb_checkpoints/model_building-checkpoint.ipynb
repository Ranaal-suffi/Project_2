Build a regression model.

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
#yelp_df is from the joining_data.ipynb
yelp_bikes_df= pd.read_json('yelp_bikes_df.json')
yelp_bikes_df.head()

# model
# y = yelp_bikes_df['available_bikes']
# x = yelp_bikes_df['rating']
# x = sm.add_constant(x)

# model = sm.OLS(y, x)
# results = model.fit()
# print(results.summary())
# using availabel_bikes as a y and rating as x which was the constant give R value of zero which means there is no correlation, same result when changing the y & x
# the p value for rating > 0.05 which makes it insgnificant
```

```python
#trying to find if there is any correlation between all the columns in the table
sns.pairplot(data=yelp_bikes_df)
plt.show()
```

```python
# it seems that there is some correlation between the rating and the resturant location when we look at the latitude and longtitude
yelp_bikes_df.corr()
```

```python
#to check if there is a signficant correlation between location and resturant rating I will use pearson's correlation coefficiant test
import scipy

stat, p1 = scipy.stats.pearsonr(yelp_bikes_df['available_bikes'], yelp_bikes_df['longtitude'])
stat, p2 = scipy.stats.pearsonr(yelp_bikes_df['free_slots'], yelp_bikes_df['latitude'])
stat, p3 = scipy.stats.pearsonr(yelp_bikes_df['rating'], yelp_bikes_df['longtitude'])
print(p1, p2, p3)
```

It shows there is a correlation between ratings and location longtitud, available_bikes and longitude, and free_slots and latitude, as the p value is < 0.05

```python
#simple linear regression using rating and available bikes
X = yelp_bikes_df['longtitude']
y = yelp_bikes_df['available_bikes']
# X = sm.add_constant(X)
```

```python
X.head
```

```python
model = sm.OLS(y, X)
results = model.fit() 
print(results.summary())
```

R^2 was 0.76 which means that longitude explains 76% of the variation in the available_bikes

```python
x = yelp_bikes_df['latitude']
y = yelp_bikes_df['free_slots']
# x = sm.add_constant(x)

model = sm.OLS(y, x)
results = model.fit()
print(results.summary())
# the p value for available bikes < 0.05 which makes it sgnificant
```

R^2 was 0.581 which means that 58% of the variance in free_slots can be explained by latitude

```python
#Multivariate Linear Regression
y = yelp_bikes_df['rating']
X = yelp_bikes_df[['latitude', 'longtitude']]
X.head()
```

```python
model = sm.OLS(y, X) 
results = model.fit() 
print(results.summary())
```

These results are very promising, R^2 of 0.994 meaning that 99.4% of the variance in rating can be explained by the longitude and latitude. This has some very promising effects from a business context

How can you turn the regression model into a classification model?

We can divide all ratings into 'high' and 'low' where high ratings are ratings >= 4 and low ratings < 4. Then we can fit a logistic regression model to the data based on latitude and longitude
